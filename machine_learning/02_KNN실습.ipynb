```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target, random_state = 2021)
```

```python
import pandas as pd
X = pd.DataFrame(iris_data.data, columns = iris_data.feature_names)
y = pd.DataFrame(iris_data.target, columns = ['label'])

data = pd.concat([X,y], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:,:-1], data.iloc[:,-1], random_state = 0)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# c: 색상을 결정
# marker: 마커모양
# his_kwds: 히스토그램 갯수를 결정
# s : 마커사이즈
# alpha: 투명도
pd.plotting.scatter_matrix(X_train, c=y_train, figsize=(15,15),
                           marker='o', hist_kwds={'bins':20}, s=60, alpha = 0.8)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors=1)

knn_model.fit(X_train,y_train)
```

```python
pred =knn_model.predict(X_test)

print(knn_model.score(X_train,y_train))
print(knn_model.score(X_test, y_test))
```

```python
train_acc = []
test_acc = []

neighbor = range(1,15,2)
for i in neighbor:
    knn_model2 = KNeighborsClassifier(n_neighbors=i)
    knn_model2.fit(X_train, y_train)
    train_acc.append(knn_model2.score(X_train,y_train))
    test_acc.append(knn_model2.score(X_test,y_test))
    
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
# random_state = 2021
plt.plot(neighbor, train_acc, label = 'Train_Acc')
plt.plot(neighbor, test_acc, label = 'Test_Acc')
plt.legend(loc = 'best')
plt.ylabel('ACC')
plt.xlabel('n_neighbor')
plt.grid()
plt.show()
```

```python
X = pd.DataFrame(cancer.data, columns= cancer.feature_names)
y = pd.DataFrame(cancer.target, columns = ['label'])

df_cancer = pd.concat([X,y],axis=1)
df_cancer.head()
```

```python
# 특성과 라벨간의 상관관계
df_cancer.corr()
```

```python
# 훈련데이터와 테스트데이터로 분리
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target,
                                                    random_state = 77, test_size = 0.3)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
knn_model3 = KNeighborsClassifier(n_neighbors=4)
knn_model3.fit(X_train,y_train)

print(knn_model3.score(X_train,y_train))
print(knn_model3.score(X_test,y_test))
```

```python
train_acc = []
test_acc = []

neighbor = range(1,30,2)
for i in neighbor:
    knn_model3 = KNeighborsClassifier(n_neighbors=i)
    knn_model3.fit(X_train, y_train)
    train_acc.append(knn_model3.score(X_train,y_train))
    test_acc.append(knn_model3.score(X_test,y_test))


plt.plot(neighbor, train_acc, label = 'Train_Acc')
plt.plot(neighbor, test_acc, label = 'Test_Acc')
plt.legend(loc = 'best')
plt.ylabel('ACC')
plt.xlabel('n_neighbor')
plt.grid()
plt.show()
```

```python
from sklearn.metrics import accuracy_score

pred = knn_model3.predict(X_test)

print(accuracy_score(pred,y_test))
```
