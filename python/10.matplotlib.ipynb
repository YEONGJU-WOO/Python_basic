## 여러 개의 plot을 그리는 방법(subplot)
- subplot(row, column, index) : 행과 열을 설정하여 그래프를 생성

```python
data = np.arange(100,201)

plt.subplot(2 ,1 ,1)
plt.plot(data)

data2 = np.arange(200,301)
plt.subplot(2 ,1 ,2)
plt.plot(data2)

plt.show()
```

```python
data = np.arange(100,201)

plt.subplot(1,3,1) #131로 대체 가능
plt.plot(data)

data2 = np.arange(200,301)
plt.subplot(1,3,2)
plt.plot(data2)

data3 = np.arange(300,401)
plt.subplot(1,3,3)
plt.plot(data3)

#그래프 간 자동 여백 설정
plt.tight_layout()
plt.show()
```

## 여러 개의 plot을 그리는 방법(subplots)
- subplots(행, 열): fig(그래프 크기)와 axes(각 그래프의 좌표) 

```python
data = np.arange(1,51)

fig, axes = plt.subplots(2,3)

#좌표별 그래프 데이터 설정
axes[0,0].plot(data)
axes[0,1].plot(data*data)
axes[0,2].plot(data**3)
axes[1,0].plot(data ** data)
axes[1,1].plot(-data)
axes[1,2].plot(data // 20)

plt.tight_layout()
plt.show()
```

## 선 투명도 설정

```python
data = np.arange(10)

#alpha : 투명도 설정
plt.plot(data, data*2, color = 'r', alpha = 0.1)
plt.plot(data, data*2+10, color = 'r', alpha = 0.3)
plt.plot(data, data*2+20, color = 'r', alpha = 0.6)
plt.plot(data, data*2+30, color = 'r', alpha = 0.8)

plt.show()
```

## Bar Chart

```python
x = ['수학', '프로그래밍', '통계', '데이터 과학', '물리']
y = [70,95,67,81,65]

plt.figure(figsize=(2,3))
plt.bar(x,y, color = 'r', alpha=0.2)
plt.title('과목')
plt.xticks(rotation=90)
plt.ylabel('점수')
plt.show()
```

```python
x = ['수학', '프로그래밍', '통계', '데이터 과학', '물리']
y = [70,95,67,81,65]

plt.figure(figsize=(5,3))
plt.barh(x,y, color = 'r', alpha=0.2)
plt.title('과목')
plt.xlabel('점수')
plt.show()
```

```python
x_label = ['수학', '프로그래밍', '통계', '데이터 과학', '물리']
x = np.arange(len(x_label))
y_1 = [70,95,67,81,65]
y_2 = [80,65,79,98,89]

width = 0.3 # bar의 너비 값

fig, axes = plt.subplots()


#axes.bar(x축 위치, y축 데이터, bar의 너비, 색상, 투명도)
axes.bar(x - width/2, y_1, width, color = 'b', alpha = 0.3)
axes.bar(x + width/2, y_2, width, color = 'r', alpha = 0.3)
axes.set_xticklabels(x_label)

plt.title('과목')
plt.ylabel('점수')
plt.xticks(x)
plt.show()
```

## Pie Chart

```python
label = ['사과', '바나나', '멜론', '포도']
ratio = [45,32,15,8]


plt.figure(figsize=(12,6))
plt.pie(ratio, labels = label, autopct='%1.1f%%')
plt.title("선호하는 과일 비율")
plt.show()
```

```python
labels = ['삼성', '화웨이', '애플', '샤오미', '기타']
ratio = [30.4, 25.7, 23.4, 15.6, 4.9]
explode = [0.1,0,0,0,0]

plt.figure(figsize=(8,4))

# autopct: pie별 비율을 표현
# explode: 각 pie별 떼어내는 정도
# startangle: pie 그래프를 그릴 때 시작하는 각도

plt.pie(ratio, labels = labels, autopct='%1.1f%%', explode = explode, shadow=True, startangle = 90)
plt.title('스마트폰 시장 점유율')
plt.show()
```

```python
labels = ['삼성', '화웨이', '애플', '샤오미', '기타']
ratio = [30.4, 25.7, 23.4, 15.6, 4.9]
explode = [0.1,0,0,0,0]
colors = ['red', 'pink', 'yellow', 'orange', 'magenta']

plt.figure(figsize=(8,4))

patchs, texts, autotexts = plt.pie(ratio, labels = labels, autopct='%1.1f%%', 
                                   explode = explode,shadow=True, startangle = 90, colors = colors)

# texts: pie별 label
# autotexts: pie별 안쪽 데이터

for text in texts:
    text.set_fontsize(20)
    text.set_color('gray')

for auto in autotexts:
    auto.set_fontsize(15)
    auto.set_color('black')


plt.title('스마트폰 시장 점유율')
plt.show()
```
